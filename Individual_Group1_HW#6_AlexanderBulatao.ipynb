## Proposition 2: Find emails of employees or job candidates[](https://jupyter.org/try-jupyter/notebooks/?path=Individual_Group1_HW%236_AlexanderBulatao.ipynb#Proposition-2:-Find-emails-of-employees-or-job-candidates)

### Functional Specification:  Use subqueries to filter emails for employees and job candidates from HumanResources. Use Person.EmailAddress linked by BusinessEntityID. Use _UNION_ clause to remove duplicates and combine the results.[](https://jupyter.org/try-jupyter/notebooks/?path=Individual_Group1_HW%236_AlexanderBulatao.ipynb#Functional-Specification:%C2%A0-Use-subqueries-to-filter-emails-for-employees-and-job-candidates-from-HumanResources.-Use-Person.EmailAddress-linked-by-BusinessEntityID.-Use-UNION-clause-to-remove-duplicates-and-combine-the-results.)

### NACE Competencies: _Communication_: Enables efficient outreach for HR or internal updates

```sql
SELECT EmailAddress
FROM Person.EmailAddress
WHERE BusinessEntityID IN (SELECT BusinessEntityID FROM HumanResources.Employee)
UNION
SELECT EmailAddress
FROM Person.EmailAddress
WHERE BusinessEntityID IN (SELECT BusinessEntityID FROM HumanResources.JobCandidate)
```

## Proposition 3: Find emails of employees AND job candidates[](https://jupyter.org/try-jupyter/notebooks/?path=Individual_Group1_HW%236_AlexanderBulatao.ipynb#Proposition-3:-Find-emails-of-employees-AND-job-candidates)

### Functional Specification:  Use subqueries to filter emails for employees and job candidates from HumanResources. Use Person.EmailAddress linked by BusinessEntityID. Use _UNION ALL_ clause to combine results, with duplicates[](https://jupyter.org/try-jupyter/notebooks/?path=Individual_Group1_HW%236_AlexanderBulatao.ipynb#Functional-Specification:%C2%A0-Use-subqueries-to-filter-emails-for-employees-and-job-candidates-from-HumanResources.-Use-Person.EmailAddress-linked-by-BusinessEntityID.-Use-UNION-ALL-clause-to-combine-results,-with-duplicates)

### Nace Competencies: _Technology_: Develops the ability to be familiar with certain techniques for efficiency, in this case handling duplicates

```sql
SELECT EmailAddress
FROM Person.EmailAddress
WHERE BusinessEntityID IN (SELECT BusinessEntityID FROM HumanResources.Employee)
UNION ALL
SELECT EmailAddress
FROM Person.EmailAddress
WHERE BusinessEntityID IN (SELECT BusinessEntityID FROM HumanResources.JobCandidate)
```
